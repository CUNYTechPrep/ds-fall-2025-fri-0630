### Clean the Data

Well, we can’t get our answers with what we currently have, so you’ll have to make some changes.

Let’s not worry about anything below the first 5 values for now. Convert the top 5 to a single canonical value―say, `'US'`, which is nice and short.

```python
# Replace them all with 'US'.
df_salary_tech  = df_salary_tech.replace({'United States': 'US',
    'USA': 'US',
    'US' :'US',
    'U.S.': 'US',
    'United States of America' : 'US',})
```

Have a look at the count of each unique country again now.

```python
# Count again.
print(df_salary_tech['country'].value_counts())


```

Did you notice anything interesting?

```python
# BONUS CREDIT: resolve [most of] those anomalous cases too without exhaustively taking every variant literally into account.

ignorethisvariable = r""" 
some notes: 
i asked chatgpt for regex help but it kept on focusing on individual cases, which i wanted to avoid, so i learned regex out of spite
    - what is this???? and they said ai was smart: 
        - us_pattern = r'^(us|usa|u\.s\.a?\.?|united\s+states?|united\s+states\s+of\s+america|unitedstates|america|state|us of us|usds|usd|unite uss|uniyed uss|unites uss|usa )$'
so here is what i learned:
- there are two outliers: "ISA", "ss" and "United Stares", so I hardcoded them in
- Australia was an issue b/c it contains "us", so i included "\bus.*" This prevents any match that does not start with "us" 
- I noticed a few typos being in the word "united" but not in "state(s)" so i used ".*state.*"
    - ex: 'United State of America', 'Uniyed states ', 'Unite States', 'UnitedStates'
- I included the strip because there were some outliers having spaces, " US" and "US "
- the ignorecase flag is cool since a lot of differences were with cases but not spelling
"""

us_pattern = r'(.*america.*|.*state.*|\bus.*|usa|u\.s?\.?a?\.?|ISA|ss|United Stares)'

df_salary_tech['country'] = df_salary_tech['country'].str.strip()

df_salary_tech['country'] = df_salary_tech['country'].str.replace(
    us_pattern, 'US', flags=re.IGNORECASE, regex=True
)

```

### Bonus #0

Apply the same steps as before to `df_salary`, but do not filter for any specific industry. Do perform the other data cleaning stuff, and get to a point where you can generate the minimum, mean, and maximum by state.

```python
#filter USD
select_condition = df_salary['currency'] == 'USD'
df_salary = df_salary[select_condition].copy()

#clean US duplicates
df_salary  = df_salary.replace({'United States': 'US',
    'USA': 'US',
    'US' :'US',
    'U.S.': 'US',
    'United States of America' : 'US',})
us_pattern = r'(.*america.*|.*state.*|\bus.*|usa|u\.s?\.?a?\.?|ISA|ss|United Stares)'
df_salary['country'] = df_salary['country'].str.strip()
df_salary['country'] = df_salary['country'].str.replace(
    us_pattern, 'US', flags=re.IGNORECASE, regex=True
)

#fix salary
df_salary['salary'] = pd.to_numeric(df_salary['salary'],  downcast="float", errors = 'coerce')
df_salary.dropna(subset=["salary"], inplace=True)
df_state_salary =  df_salary.groupby('state')['salary'].agg(['min', 'mean', 'max'])
df_state_salary.sort_values(by='mean', ascending=False, inplace=True)
df_state_salary.head()
```

### Bonus #3

Show the quantiles instead of just minimum, mean, and maximum―say 0%, 5%, 25%, 50%, 75%, 95%, and 100%. Outliers may be deceiving.

Sort by whatever interests you―like say the *50th* percentile.

And throw in a count by state too. It would be interesting to know how many data points contribute to the figures for each state. (*Hint: your nice formatting from Bonus #1 might not work this time around.* 😜)

```python
#seperate states 
df_salary = df_salary[df_salary['state'].str.count(',') == 0]

#use quantile function to well, find quantiles
quantiles = [0, 0.05, 0.25, 0.5, 0.75, 0.95, 1]
df_state_quantiles = df_salary.groupby('state')["salary"].quantile(quantiles).unstack()

#find and join counts for each state then append
df_state_counts = df_salary.groupby('state')['salary'].count()
df_join = df_state_quantiles.join(df_state_counts.rename("count"))

#sort by 75th percentile and apply style
df_join = df_join.sort_values([0.75], ascending=False)
df_join.style.format('${:,.2f}', subset=quantiles)

```
