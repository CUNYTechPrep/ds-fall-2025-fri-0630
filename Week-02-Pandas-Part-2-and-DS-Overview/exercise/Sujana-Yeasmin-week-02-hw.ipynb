### Clean the Data

Well, we can’t get our answers with what we currently have, so you’ll have to make some changes.

Let’s not worry about anything below the first 5 values for now. Convert the top 5 to a single canonical value―say, `'US'`, which is nice and short.

```python
# Replace them all with 'US'.

tech_df = df_salary_USD[df_salary_USD['Industry'] == 'Computing or Tech'].copy()

top_5_countries = tech_df['Country'].value_counts().head(5).index.tolist()
print("Top 5 countries in tech:", top_5_countries)

tech_df['Country'] = tech_df['Country'].replace(top_5_countries, 'US')

tech_df

```

It’s looking good so far. Let’s find out the minimum, mean, and maximum (in that order) salary by state, sorted by the mean in descending order.

```python
# Find the minimum, mean, and maximum salary in USD by U.S. state.
us_tech_df = tech_df[tech_df['Country'] == 'US'].copy()

us_tech_df = us_tech_df.dropna(subset=['Salary'])
```

Well, pooh! We forgot that `salary` isn’t numeric. Something wrong must be fixed.

```python
# Fix it.
us_tech_df = tech_df[tech_df['Country'] == 'US'].copy()
us_tech_df['Salary'] = pd.to_numeric(us_tech_df['Salary'], errors='coerce')

us_tech_df = us_tech_df.dropna(subset=['Salary'])
us_tech_df.dtypes
```

Let’s try that again.

```python
# Try it again. Yeah!

salary_stats = us_tech_df.groupby('City in US')['Salary'].agg(['min', 'mean', 'max'])

print(salary_stats)
```

That did the trick! Now let’s narrow this to data 2021 and 2022 just because (lel). *(Hint: that timestamp column may not be a temporal type right now.)*

```python
#Timestamp is currently an Object. Need to conver it to daytime 

us_tech_df['Timestamp'] = pd.to_datetime(us_tech_df['Timestamp'])
```

```python
# Filter the data to within 2021, 2022, or 2023, saving the DataFrame to a new variable, and generate the summary again.

us_tech_df['Year'] = us_tech_df['Timestamp'].dt.year
us_tech_recent = us_tech_df[us_tech_df['Year'].isin([2021, 2022])]

salary_stats_years = us_tech_recent.groupby('City in US')['Salary'].agg(['min', 'mean', 'max'])

print(salary_stats_years)
```
