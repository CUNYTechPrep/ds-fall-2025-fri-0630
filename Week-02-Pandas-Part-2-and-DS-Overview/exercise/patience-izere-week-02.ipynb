### Bonus #0

Apply the same steps as before to `df_salary`, but do not filter for any specific industry. Do perform the other data cleaning stuff, and get to a point where you can generate the minimum, mean, and maximum by state.

```python
import pandas as pd

# 1. Start with df_salary (all industries, renamed columns)
df = df_salary.copy()

# 2. Coerce salary ‚Üí numeric; drop invalid
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
df = df.dropna(subset=['salary'])

# 3. Restrict to USD (we still haven‚Äôt converted EUR/GBP‚Ä¶)
df = df[df['currency'] == 'USD']

# 4. Parse timestamps and drop bad rows
df['timestamp'] = pd.to_datetime(df['timestamp'], infer_datetime_format=True, errors='coerce')
df = df.dropna(subset=['timestamp'])

# 5. Now group by state and compute min/mean/max
state_summary = (
    df.groupby('state')['salary'].agg(min='min', mean='mean', max='max')
      .sort_values('mean', ascending=False)
)

state_summary
```

### Bonus #2

Filter out the non-single-states (e.g., `'California, Colorado'`) in the most elegant way possible (i.e., *not* by blacklisting all the bad values).

```python
# Filter to rows where state does NOT contain a comma
df_single_state = df[~df['state'].str.contains(',', na=False)].copy()

# Sanity check: no commas anywhere
assert not df_single_state['state'].str.contains(',', na=False).any()

# Recompute the summary
single_state_summary = (
    df_single_state.groupby('state')['salary'].agg(min='min', mean='mean', max='max')
      .sort_values('mean', ascending=False)
)

single_state_summary
```

### Bonus #3

Show the quantiles instead of just minimum, mean, and maximum‚Äïsay 0%, 5%, 25%, 50%, 75%, 95%, and 100%. Outliers may be deceiving.

Sort by whatever interests you‚Äïlike say the *50th* percentile.

And throw in a count by state too. It would be interesting to know how many data points contribute to the figures for each state. (*Hint: your nice formatting from Bonus #1 might not work this time around.* üòú)

```python
# Define desired quantiles
qs = [0, .05, .25, .5, .75, .95, 1.0]

# Group and aggregate
quantile_summary = (
    df_single_state.groupby('state')['salary']
      .agg(
         count='count',
         q0=lambda x: x.quantile(qs[0]),
         q5=lambda x: x.quantile(qs[1]),
         q25=lambda x: x.quantile(qs[2]),
         q50=lambda x: x.quantile(qs[3]),
         q75=lambda x: x.quantile(qs[4]),
         q95=lambda x: x.quantile(qs[5]),
         q100=lambda x: x.quantile(qs[6]),
      ).sort_values('q50', ascending=False)  # sort by median
)

quantile_summary
```
